handling changing the shape of tensors:

## Changing the Number of Dimensions
passing a single instance of input to your model. PyTorch models generally expect *batches* of input. 

ex: model working with 3 x 266 x 266 images - a 226-pixel square with 3 color channels. when loading and transforming it, we'll get a tensor of shape ```(3, 226, 226)```. The model expects an input shape of ```(N, 3, 226, 226)```, where ```N``` is the number of images in the batch. So how do you make a batch of one?

````python
import torch
````

````python
a = torch.rand(3, 226, 226)
b = a.unsqueeze(0)

print(a.shape)
print(b.shape)
````

```unsqueeze()``` adds a dimension of extent 1. ```unsqueeze(0)``` adds it as a new zeroth dimension (we now have batch size of 1)

what is squeezing? We take advantage of the fact that any dimension of extent 1 *does not* change the number of elements in the tensor. 

````python
c = torch.rand(1, 1, 1, 1, 1)
print(c)
````

# Moving to GPU 
Acceleration is possible on CUDA-compatible Nvidia GPUs. *CUDA* means *Compute Unified Device Architecture* which is Nvidia's platform for parallel computing. 

First we check if there is a GPU available, with ```is_available()``` method

````python
import torch
````

````python
if torch.cuda.is_available():
    print('On GPU')
else:
    print('Sorry, CPU only')
````
